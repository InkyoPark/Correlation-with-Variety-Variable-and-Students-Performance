```python
# csv읽기 위한 라이브러리
import pandas as pd
# 시각화를 위한 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data_set읽기
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
# 개요파악
df.describe()
```

```python
# 수학,읽기,쓰기 간의 관계를 시각화
sns.pairplot(df[["math score", "reading score", "writing score"]], height=5)
```

```python
# 세점수의 평균값을 적용시킴 
def average_score(df):
    return (df["math score"] + df["reading score"] + df["writing score"])/3

df["average score"] = df.apply(average_score, axis=1)
```

```python
# check the data
df.head(5)
```

```python
# "점심"과 "수학점수"
sns.catplot(x="lunch", y="math score", hue="gender", kind="boxen", data=df, height= 15, palette= sns.color_palette(["red", "blue"]))
plt.title("math")

# "점심"과 "읽기점수"
sns.catplot(x="lunch", y="reading score", hue="gender", kind="boxen", data=df, height= 15, palette= sns.color_palette(["red", "blue"]))
plt.title("reading")

# "점심"과 "쓰기점수"
sns.catplot(x="lunch", y="writing score", hue="gender", kind="boxen", data=df, height= 15, palette= sns.color_palette(["red", "blue"]))
plt.title("writing")
```

```python
sns.catplot(x="lunch", y="average score", hue="gender", kind="boxen", data=df, height= 15, palette= sns.color_palette(["red", "blue"]))
plt.title("average")
```

```python
# "시험준비코스"와 "평균점수"
sns.catplot(x="test preparation course", y="average score", data=df, hue="gender", kind="boxen", height= 15, 
            palette= sns.color_palette(["red", "blue"]))
plt.title("average")
```

```python
# "부모의 최종학력"과 "평균점수"
sns.catplot(x="parental level of education", y="average score", data=df, kind="boxen", height= 15)
plt.title("average")
```

```python
# "인종"과 "평균점수" 
sns.catplot(x="race/ethnicity", y="average score", data=df, hue="gender", kind="boxen", height= 15, 
            palette= sns.color_palette(["red", "blue"]))
plt.title("average")
```
